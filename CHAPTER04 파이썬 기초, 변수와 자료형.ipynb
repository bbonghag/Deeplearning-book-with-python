### 4.4.2 if 문

- if 조건식이 True라면 : ~을 실행

```python
n = 2
if n == 2:
    print(n)
```

```python
n = 16
if n > 15:
    print('큰 숫자')
```

### 4.4.3 else
- if문에 else를 쓰면 '그렇지 않으면 ~한다' 로 다른 조건문이 실행된다

```python
n = 2
if n == 1:
    print(n)
else:
    print('유감')
```

### 4.4.4 elif

- if 문의 조건이 성립되지 않았을 경우 다른 조건을 정의하고 싶을 때 elif를 사용한다
- 여러개 사용가능

```python
n = 2
if n == 1:
    print('금메달')
elif n == 2:
    print('은메달')
elif n == 3:
    print('동메달')
else:
    print('유감')
```

- 윤년 여부 계산문제
- 연도가 400으로 나누어떨어지지 않고, 100으로 나누어 떨어지면 윤년이 아니다 = 평년
- 위와 같지 않을 때 연도가 4로 나누어떨어지면 윤년이다
- 윤년이 아니면 평년
- 윤년의 경우 '-년은 윤년이다'
- 평년의 경우 '-년은 평년이다'

```python
year = int(input())

if year % 400 != 0 and year % 100 == 0:
    print(str(year) + '년은 평년이다')
elif year % 4 == 0:
    print(str(year) + '년은 윤년이다')
else:
    print(str(year) + '년은 평년이다')
```
