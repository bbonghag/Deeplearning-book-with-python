# CHAPTER11 matplotlib 사용하기

## 11.1 한 종류의 데이터 시각화하기

### 11.1.1 그래프에 데이터 표시하기



```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

### 11.1.2 그래프의 표시 범위 설정하기

- 맷플로립으로 그래프를 그릴때 그래프의 표시 범위가 자동으로 설정된다
- 각 축에 할당된 데이터의 최소, 최댓값이 표시범위가 된다
- 임의로 범위를 지정 가능하다

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.ylim(0,1)
plt.plot(x,y)
plt.show()
```

### 11.1.3 그래프 요소에 이름 설정하기

- 그래프와 x,y축에 이름을 설정할 수 있다

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('sin function')
plt.plot(x,y)
plt.show()
```

### 11.1.4 그래프에 그리드 표시하기

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.grid(True)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('sin function')
plt.plot(x,y)
plt.show()
```

### 11.1.5 그래프 축에 눈금 설정하기

- 데이터의 종류에 따라 자동으로 붙은 눈금이 맞지 않을 때 임의로 설정가능하다
- plt.xticks(눈금 삽입 위치, 삽입할 눈금)

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.grid(True)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('sin function')

positions = [0,np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['0', '90', '180', '270', '360']
plt.xticks(positions, labels)
plt.plot(x,y)
plt.show()
```

## 11.2 여러 데이터 시각화하기(1)

### 11.2.1 하나의 그래프에 두 종류의 데이터 표시하기

- 여러개의 데이터를 표시할 경우 색상으로 구분하고 라벨을 붙인다

```python
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.grid(True)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('sin_cos function')

positions = [0,np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['0', '90', '180', '270', '360']
plt.xticks(positions, labels)
plt.plot(x,y1,color='k')
plt.plot(x,y2,color='b')
plt.show()
```

### 11.2.2 계열에 라벨 설정하기

- 그래프에서 각 선이 나타내는 것이 무엇인지 주석을 달아줘서 구별할 수 있도록 해준다

```python
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)

plt.grid(True)
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.title('sin_cos function')

positions = [0,np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['0', '90', '180', '270', '360']
plt.xticks(positions, labels)
plt.plot(x,y1,color='k', label='sin')
plt.plot(x,y2,color='b', label='cos')
plt.legend()
# plt.legend(['sin', 'cos'])
plt.show()
```

## 11.3 여러 데이터 시각화하기(2)

### 11.3.1 그림 크기 설정하기

- 그래프를 그릴 때 가로,세로 크기를 임의로 설정 가능하다

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.show()
```

### 11.3.2 서브플롯 만들기

- 작은 서브플롯 단위로 그래프를 그리는 영역을 지정할 수 있다


```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(2,3,5)
plt.plot(x,y)

axi = []
for i in range(6):
    if i == 4:
        continue # (2,3,5)에 이미 그래프가 있으므로 패스.
    fig.add_subplot(2,3,i+1)

plt.show()
```

### 11.3.3 서브플롯 주변의 여백 조정하기

- 서브플롯을 나란히 배치한 경우 축 이름과 제목을 설정하면 서브플롯끼리 겹치기도 한다
- 주변의 여백을 조정해서 겹치는걸 방지할 수 있다

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
positions = [0,np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['0', '90', '180', '270', '360']
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(2,3,5)
plt.subplots_adjust(wspace=1, hspace=1)

ax.plot(x,y)

axi = []
for i in range(6):
    if i == 4:
        continue 
    fig.add_subplot(2,3,i+1)

plt.show()
```

### 11.3.4 서브플롯의 그래프 표시 범위 설정하기

- 그래프의 표시 범위를 x,y축에 대해 각각 설정할 수 있다
- x축의 표시 범위 설정 : ax.set_xlim(범위)
- y축의 표시 범위 설정 : ax.set_ylim(범위)
- ax는 서브플롯 객체 

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['90', '180', '270', '360']


fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(2,3,5)
plt.subplots_adjust(wspace=1, hspace=1)
ax.set_ylim([0,1])


ax.plot(x,y)

axi = []
for i in range(6):
    if i == 4:
        continue 
    fig.add_subplot(2,3,i+1)

plt.show()
```

### 11.3.5 서브플롯의 그래프 요소에 이름 붙이기

- 그래프의 제목 설정 : ax.set_title('제목')
- x축 이름 설정 : ax.set_xlabel('x축 이름')
- y축 이름 설정 : ax.set_ylabel('y축 이름')

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['90', '180', '270', '360']


fig = plt.figure(figsize=(9,6)) # figure 객체 만들기
ax = fig.add_subplot(2,3,5) # 2x3 레이아웃에서 위에서 두번째 행, 왼쪽에서 두번째 열에 서브플롯 객체 ax를 만듬
plt.subplots_adjust(wspace=1, hspace=1) # 그림 내 서브플롯 간격을 가로세로 1.0으로 설정

ax.set_title('y=sin(x)')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')


ax.plot(x,y)

# 빈 공간을 서브플롯으로 채움
axi = []
for i in range(6):
    if i == 4:
        continue 
    fig.add_subplot(2,3,i+1)

plt.show()
```

### 11.3.6 서브플롯 그래프에 그리드 표시하기

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['90', '180', '270', '360']


fig = plt.figure(figsize=(9,6)) 
ax = fig.add_subplot(2,3,5) 
plt.subplots_adjust(wspace=1, hspace=1) 

ax.grid(True)

ax.set_title('y=sin(x)')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

ax.plot(x,y) # x,y를 그래프에 플롯

# 빈 공간을 서브플롯으로 채움
axi = []
for i in range(6):
    if i == 4:
        continue 
    fig.add_subplot(2,3,i+1)

plt.show()
```

### 11.3.7 서브플롯 그래프 축에 눈금 설정하기

- 이전에 그래프 축 눈금 설정한 것 처럼 서브플롯 그래프도 눈금이 설정가능하다
- 전에는 한번에 설정한 것과 달리 이번에는 따로따로 설정해준다
- ax.set_xticks('삽입 위치 리스트')
- ax.set_xtickslabes('눈금 리스트')

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)
positions = [np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ['90', '180', '270', '360']


fig = plt.figure(figsize=(9,6)) 
ax = fig.add_subplot(2,3,5) 
plt.subplots_adjust(wspace=1, hspace=1) 

ax.grid(True)

ax.set_title('y=sin(x)')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

# 서브플롯 ax 그래프의 x축에 눈금 설정
ax.set_xticks(positions)
ax.set_xticklabels(labels)

ax.plot(x,y) 

axi = []
for i in range(6):
    if i == 4:
        continue 
    fig.add_subplot(2,3,i+1)

plt.show()
```
