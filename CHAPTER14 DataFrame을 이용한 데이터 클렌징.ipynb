## 14.3 결측치

### 14.3.1 리스트와이즈 삭제와 페어와이즈 삭제

- 데이터가 누락된 행을 통째로 지우는 것을 리스트와이즈 삭제라고 한다
- 결손이 적은 열만 남기는 것을 페어와이즈 삭제라 한다

```python
from numpy import nan as NA
import numpy as np


sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA
sample_data_frame
```

```python
sample_data_frame.dropna()
```

```python
sample_data_frame[[0,1,2]].dropna()
```

```python
np.random.seed(0)

sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

sample_data_frame[[0,2]].dropna() # 원하는 열만 남김
```

### 14.3.2 결측치 보완

- drop으로 결측치를 삭제하는 방법도 있지만 대체할 값을 넣는 방법도 있다

```python
sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

sample_data_frame.fillna(0)
```

```python
sample_data_frame.fillna(method='ffill')
```

```python
sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[6:,2] = NA


sample_data_frame.fillna(method='ffill')
```

### 14.3.3 결측치 보완(평균값 대입법)

- 결측치를 열, 행의 평균값으로 대체하기도 한다

```python
sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[2,2] = NA
sample_data_frame.iloc[5:,3] = NA

sample_data_frame.fillna(sample_data_frame.mean())
```

```python
np.random.seed(0)

sample_data_frame = pd.DataFrame(np.random.rand(10,4))

sample_data_frame.iloc[1,0] = NA
sample_data_frame.iloc[6:,2] = NA

sample_data_frame.fillna(sample_data_frame.mean())
```

### 14.4.3 매핑

- 매핑은 공통의 키 역할을 하는 데이터의 값을 가져오는 처리

```python
attri_data1 = {'ID' : ['100', '101', '102', '103', '104', '106', '108', '110', '111', '113'],
               'city':['서울', '부산', '대전', '광주', '서울', '서울', '부산', '대전', '광주', '서울'],
               'birth_year':[1990,1989,1992,1997,1982,1991,1988,1990,1995,1981],
               'name':['영이', '순돌', '짱구', '태양', '션', '유리', '현아', '태식', '민수', '호식']}

attri_df = pd.DataFrame(attri_data1)
attri_df
```

- 위의 attri_df의 city컬럼을 기반으로 대응하는 지역명을 새 컬럼으로 추가하는 것을 매핑 처리라 한다

```python
city_map = {'서울':'서울',
            '광주':'전라도',
            '부산':'경상도',
            '대전':'충청도'}
attri_df['region'] = attri_df['city'].map(city_map)
attri_df
```

```python
MS = {'서울':'중부',
      '대전':'중부',
      '광주':'남부',
      '부산':'남부'
      } 

attri_df['MS'] = attri_df['city'].map(MS)
attri_df
```

### 14.4.4 구간 분할
- 이산 범위의 데이터를 분할하여 집계할 경우 사용하는 기능
- 구간을 분할할 리스트를 준비하고, pandas의 cut()함수로 처리한다

```python
attri_data1 = {'ID' : ['100', '101', '102', '103', '104', '106', '108', '110', '111', '113'],
               'city':['서울', '부산', '대전', '광주', '서울', '서울', '부산', '대전', '광주', '서울'],
               'birth_year':[1990,1989,1992,1997,1982,1991,1988,1990,1995,1981],
               'name':['영이', '순돌', '짱구', '태양', '션', '유리', '현아', '태식', '민수', '호식']}

attri_df = pd.DataFrame(attri_data1)
attri_df
```

```python
birth_year_bins = [1980,1985,1990,1995,2000]

birth_year_cut_data = pd.cut(attri_df.birth_year, birth_year_bins)
birth_year_cut_data
```

```python
pd.value_counts(birth_year_cut_data)
```

```python
group_names = ['first1980', 'second1980', 'first1990', 'second1990']
birth_year_cut_data = pd.cut(attri_df.birth_year, birth_year_bins, labels=group_names)
pd.value_counts(birth_year_cut_data)
```

```python
pd.cut(attri_df.birth_year, 2)
```
