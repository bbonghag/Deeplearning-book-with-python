## 5.3 while문


### 5.3.1 while문(1)

- 조건식이 False 될 때까지 무한 루프

```python
n = 2
while n > 0:
    print(n)
    n -= 1
```

### 5.3.2 while문(2)

- 조건식의 변숫값을 갱신하지 않거나 조건식이 항상 성립되로록 하면 루프가 무한정 반복되므로 주의할 것

```python
x = 5
while x != 0:
    x -= 1
    print(x)
```

### 5.3.3 while과 if

```python
x = 5
while x != 0:
    x -= 1
    if x !=0:
        print(x)
    else:
        print('Bang')
```

### 5.4 for문

### 5.4.1 for문

- for 변수 in 데이터:


```python
for i in range(5):
    print(i)
```

### 5.4.2 break

- break를 통해 반복 처리를 종료할 수 있다

```python
storages = range(1,11)
for n in storages:
    print(n)
    if n >= 5:
        break
```

```python
a = range(1,7)
for n in a:
    print(n)
    if n == 4:
        break
```

### 5.4.3 continue

- 특정 조건일 때 루프를 한번 건너뛴다

```python
a = range(1,4)
for i in a:
    if i == 2:
        continue
    print(i)
```

```python
dd = range(1,7)
for i in dd:
    if i % 2 == 0:
        continue
    print(i)
```

### 5.5.3 딕셔너리형의 루프
- 딕셔너리형의 루프에서는 키와 값을 모두 변수로 하여 반복할 수 있다
- items()를 이용해 묶은다음 두 변수에 할당한다

```python
dic
```

```python
for key, value in dic.items():
    print(key, end=' : '), print(value)
```

```python
items = {'지우개':[100,2], '펜':[200,3], '노트':[400,5]}
total_price = 0
money = int(input())

for item, price_count in items.items():
    print(f'{item}은 한 개에 {price_count[0]}원이며, {price_count[1]}개 구입합니다')
    total_price += price_count[0] * price_count[1]

print(f'지불해야 할 금액은 {total_price}입니다')

if money > total_price:
    print(f'거스름돈은 {money - total_price}입니다')
elif money == total_price:
    print('거스름돈이 없습니다')
else:
    print('돈이 부족합니다')
```
